### Relation Features | Target

#### Separation en sous-ensemble de nos données

```python
df_positif = df[df['SARS-Cov-2 exam result'] == 'positive']
```

```python
df_negative = df[df['SARS-Cov-2 exam result'] == 'negative']
```

```python
print("Dimension Resultat Positif :  ",df_positif.shape)
print("Dimension Resultat Negatif :  ",df_negative.shape)
```

### Technique de selection rapide

#### Ensemble Test | viral

```python
missing_rate = df.isna().sum() / df.shape[0]
```

```python
blood_columns = df.columns[(missing_rate < 0.9) & (missing_rate > 0.88)]
```

```python
viral_columns = df.columns[(missing_rate < 0.88) & (missing_rate > 0.75)]
```

#### Relation Target ||| Test

```python
for col in blood_columns:
    plt.figure()
    sns.distplot(df_positif[col], label='positif')
    sns.distplot(df_negative[col], label='negative')
    plt.legend()
```

##### Note sur le graphique :
     - les taux de Monocytes , Platelets, Leukocytes peut êtres liée au COVID-19
